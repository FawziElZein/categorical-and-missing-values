## Part 2:

### Cleanup: Deleting the rows or columns

```python
import numpy as np
missing_percentage = 0.3 #raised the missing percentage from 10% to 30%
list_cols = ['id', 'bathrooms', 'floors', 'lat','long']
df_ = df[list_cols].copy()
```

```python
np.random.rand(*df_.shape)
mask = np.random.rand(*df_.shape) < missing_percentage
df_[mask] = np.nan
df_
```

```python
#delete rows that have NaN values at ANY column of the data set
df_cleaned = df_.dropna()
df_cleaned
```

```python
#delete rows that have NaN values at specific columns of the data set ('bathrooms',floors')
df_cleaned = df_.dropna(subset=['bathrooms', 'floors'])
df_cleaned
```

### Cleanup: Replacing the missing values by either the median, mean, mode

```python
df_.median()
```

```python
df_.head()
```

```python
df_filled_with_median = df_.fillna(df_.median())
df_filled_with_median.head()
```
